## Suite à un partenariat entre une compagnie d'assurance et le videos club, 3000 nouveaux clients doivent être ajouter à la base de données. Toute les données necessaire pour ajouter ces nouveaux clients sont contenues dans un fichier Excel

On va donc utiliser la méthode pd.read_excel()

```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("mysql+pymysql://rafik:simplon@localhost/sakila")
# https://docs.sqlalchemy.org/en/14/core/engines.html

customers_3000 = pd.read_excel('Jeux_de_donnees\\Companie_des_3000.xlsx')
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html
```

```python
customers_3000
```

Les clients étant tous français leurs villes n'aparaissent pas dans la tables city de la base Sakila qui ne contient que des villes Américaine. Il va donc falloir les ajouter.

Pour ce faire on va créer un DataFrame à l'image de la table city contenant nos nouvelle villes :
- On va dabord récupérer le nom des colonnes de la table.
- Les villes étant toute Française, on va récupérer l'id de la France pour remplir la colonne country_id
- Puis on va remplir la colonne last_update avec la date du jour

```python
column_city = pd.read_sql_query("select * from city limit 1", engine).columns
column_city
```

```python
city = pd.DataFrame(columns = column_city)
city['city'] = customers_3000['City']
city.drop_duplicates(inplace=True)
```

```python
city
```

```python
france_id = pd.read_sql_query("select country_id from country where country = 'France'", engine).values
```

```python
france_id
```

```python
city['country_id'] = france_id[0,0]
city
```

```python
city['last_update']=pd.Timestamp.now()
city
```

```python
city.to_sql('city', if_exists='append', con=engine, index=False)
```
