# Créer son premier graph

## 1. Importer les données et filtrer les valeurs vides

```python
import pandas as pd 

donnees = pd.read_csv("jeux_de_donnees/120-years-of-olympic-history-athletes-and-results.csv", index_col=[1])
donnees
```

```python
donnees.isnull().sum()
```

On utilise la méthode fillna() de Pandas sur la colonne Medal de notre DataFrame stocké dans la variable donnees pour remplacer les NaN de cette colonne par la string "None"

```python
donnees["Medal"] = donnees["Medal"].fillna("None")
donnees
```

Ensuite, trois autres variable contiennet de valeurs manquantes : Age, Height, Weight, nous avons deux solutions ici :
- Supprimer les athlètes qui ont une valeur manquante dns au moins une de ces variable
- Calculer la moyenne de chaque variable sur l'ensemble de individus du tableau et remplacer les valeurs NaN de chaque variable par la moyenne de la variable

```python
# Création du nouveau dataset sans les lignes contenant les valeurs manquantes
donnees_NaN_supprimees = donnees.dropna()
donnees_NaN_supprimees = donnees_NaN_supprimees.drop_duplicates()
print(donnees_NaN_supprimees.isnull().sum())
```

```python
# Création du nouveau dataset avec remplacement des variables par la moyenne
moyennes = donnees[["Age", "Height", "Weight"]].mean()
donnees[["Age", "Height", "Weight"]] = donnees[["Age", "Height", "Weight"]].fillna(moyennes)
donnees_NaN_remplacees = donnees.drop_duplicates()
print(donnees_NaN_remplacees.isnull().sum())
print()
print(moyennes)
```

```python
print("Tableau avec NaN remplacés :")
print(donnees_NaN_remplacees.shape)
print("Tableau avec NaN supprimés :")
print(donnees_NaN_supprimees.shape)
```

## 2. Créer un nuage de points



```python
import matplotlib.pyplot as plt 

# On utilise la fonction figure(), qui permet de créer une nouvelle figure, vierge
plt.figure()
# La commande subplot(111) permet de créer une grille de 1*1 puis de placer le graphique à la position 1
plt.subplot(111)
# La focntion plot crée les objets axis et trace les points, en prenant les vlaeurs qu'on lui donne en x et y
plt.plot(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Weight"])
# la fonction show() permet d'afficher le graphique courant
plt.show()
```

```python
plt.figure()
plt.plot(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Weight"], marker='x', linestyle='', color = 'k', markersize="2")
plt.show()
```

Avec ce premier graph, on est capable de dire qu'il semble y avoir un lien entre la taille et le poids

## 3. Ajouter un titre principal et des labels

## 5. Tracer deux courbes

```python
plt.figure(figsize=[10,5], facecolor="lightgrey")
plt.subplot(111)
plt.plot(donnees_NaN_supprimees["Height"],donnees_NaN_supprimees["Weight"], marker="x", color="tab:blue", linestyle="")
plt.plot(donnees_NaN_supprimees["Height"],donnees_NaN_supprimees["Age"], marker="v", color="tab:green", linestyle="")

plt.title("Poinds et Age en fonction de la taille des athlète")
plt.ylabel("Poids/Age")
plt.xlabel("Taille")
plt.show()
plt.close()
```

```python
import numpy as np
age_mean = [np.mean(donnees_NaN_supprimees['Age'])] * len(donnees_NaN_supprimees['Height'])
```

```python
plt.figure(figsize=[10,5], dpi=100)
plt.subplot(111)
plt.plot(donnees_NaN_supprimees["Height"],donnees_NaN_supprimees["Weight"], marker="x", color="tab:blue", linestyle="", label="Poids")
plt.plot(donnees_NaN_supprimees["Height"],donnees_NaN_supprimees["Age"], marker="v", color="tab:green", linestyle="", label="Age")
plt.plot(donnees_NaN_supprimees["Height"], age_mean,linestyle="--", color="tab:red", label="Age moyen")

plt.title("Poinds et Age en fonction de la taille des athlète")
plt.ylabel("Poids/Age")
plt.xlabel("Taille")
plt.legend(loc="upper left")
plt.show()
plt.close()
```

## 6. Annoter un graph

```python
plt.figure(dpi=100)
plt.subplot(111)
plt.plot(donnees_NaN_supprimees["Height"],donnees_NaN_supprimees["Weight"], marker="x", color="tab:blue", linestyle="", label="Poids")
plt.plot(donnees_NaN_supprimees["Height"],donnees_NaN_supprimees["Age"], marker="v", color="tab:green", linestyle="", label="Age")
plt.plot(donnees_NaN_supprimees["Height"], age_mean,linestyle="--", color="tab:red", label="Age moyen")

plt.annotate("Moyenne = 25 ans", xy=(210,25),xytext=(250,200), arrowprops=dict(facecolor='black'))
plt.text(180,130,"y=Poids, x=Taille",size=13, bbox=dict(boxstyle="round",fc="white",ec="k"))

plt.title("Poinds et Age en fonction de la taille des athlète")
plt.ylabel("Poids/Age")
plt.xlabel("Taille")
plt.legend(loc="upper left")

plt.show()
plt.close()
```

# 4. Les différents types de graphs

## 4.1 Scatterplot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6),facecolor="white")
plt.subplot(111)
plt.scatter(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Weight"], s=0.5)
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8,6),facecolor="white")
plt.subplot(111)
colors = {'F':"tab:red", 'M':"tab:blue"}

plt.scatter(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Weight"],
            c=donnees_NaN_supprimees["Sex"].apply(lambda x: colors[x]), 
            s=0.5)
plt.show()
plt.close()
```

```python
# création d'un graphique orienté objet

fig,ax = plt.subplots(figsize=(8,6),facecolor="white")
ax.scatter(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Weight"],
            c=donnees_NaN_supprimees["Sex"].apply(lambda x: colors[x]), 
            s=0.5)
ax.set_xlabel("Taille (cm)")
ax.set_ylabel("Poids (lb)")

plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(5,10), facecolor="white")
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.scatter(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Weight"], s=0.5)
ax1.set_xlabel("Taille (cm)")
ax1.set_ylabel("Poids (lb)")

ax2.scatter(donnees_NaN_supprimees["Height"], donnees_NaN_supprimees["Age"], s=0.5)
ax2.set_xlabel("Taille (cm)")
ax2.set_ylabel("Age")

plt.show()
plt.close()
```

## 4.2 Graphique à barres

```python
print(donnees_NaN_remplacees["Team"].value_counts()[:10])

labels_x = donnees_NaN_remplacees["Team"].value_counts()[:10].keys()
valeurs_y = donnees_NaN_remplacees["Team"].value_counts()[:10].values

print(labels_x)
print(valeurs_y)
```

```python
import pandas as pd

plt.figure(facecolor="white")
plt.subplot(111)

plt.bar(labels_x,valeurs_y)

plt.xticks(labels_x, rotation=90)
plt.ylabel("Nombre d'athlètes")
plt.title("Top 10 des équipes avec le plus d'athlètes")

plt.show()
plt.close()
```
