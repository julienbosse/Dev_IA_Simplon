```python
import pandas as pd 
import matplotlib.pyplot as plt

unrate = pd.read_csv("jeux_donnees/unrate.csv")
unrate
```

## Superposition de courbes

- Créer une colonne 'MONTH' à partir de la colonne 'DATE'
- Définir la zone d'affichage à 6 pouces de large et 3 pouces de haut.
- Générer 2 courbes dans un seul graphique en utilisant la colonne 'MONTH' pour l'axe x à la place de la colonne '
    - Une courbe avec les valeurs de 1948 de couleur 'red'
    - Une courbe avec les valeurs de 1949 de couleur 'blue'
- Afficher les courbes.

```python
unrate['MONTH'] = unrate['DATE'].dt.month

plt.figure(figsize=(6,3), facecolor="white")

plt.plot(unrate['MONTH'][:12], unrate['UNRATE'][:12], color='tab:red')
plt.plot(unrate['MONTH'][:12], unrate['UNRATE'][12:24], color='tab:blue')

plt.xlabel("Mois")
plt.ylabel("Taux de hômage(%)")

plt.show()
plt.close()
```

## Ajouter plus de courbes

- Définir la zone d'affichage du graphique à 10 pouces de large et 6 pouces de haut.
- Générer les courbes suivantes sur le même graphique:
    - 1948: courbe en rouge ' red
    - 1949: courbe en bleu ' blue
    - 1950: courbe en vert 'green'
    - 1951: courbe en orange 'orange'
    - 1952: courbe en noir 'black'
- Afficher les courbes.

(Une boucle for peut être utile.)

```python
plt.figure(figsize=(10,6), facecolor="white")

for i, c in zip(range(5), ["tab:red","tab:blue","tab:green","tab:orange","black"]):

    plt.plot(unrate['MONTH'][:12], unrate['UNRATE'][12*i:12*(i+1)], color=c, label=str(1948+i))

    
plt.xlabel("Mois")
plt.ylabel("Taux de chômage(%)")
plt.title("Evolution du taux de chômage de 1948 à 1952")

plt.legend(loc="upper left")
plt.show()
plt.close()
```
