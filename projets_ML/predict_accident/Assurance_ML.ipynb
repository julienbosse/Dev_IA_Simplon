```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
```

```python

cat_attribs = ["lum", "agg", "atm", "cat_time", "journee", "vac"]

cat_attribs2 = ["lum", "atm", "cat_time", "journee", "vac"]

cat_attribs3 = ["lum", "atm", "agg", "journee", "vac"]

cat_attribs4 = ["agg"]

full_pipeline2 = ColumnTransformer([
        ("cat", OneHotEncoder(), cat_attribs),
    ])

full_pipeline3 = ColumnTransformer([
    ("cat", OneHotEncoder(), cat_attribs2),
    
])

full_pipeline4 = ColumnTransformer([
    ("cat", OneHotEncoder(), cat_attribs3),
])

full_pipeline5 = ColumnTransformer([
    ("cat", OneHotEncoder(), cat_attribs4),
])

```

```python
def learning(full_pipeline):
    ac_score = []

    skf = StratifiedKFold(n_splits=100) # stratified leave-k-out cross validation

    for train_index, test_index in skf.split(vrai_data, vrai_data['grav_class']):

        data_train, data_test = vrai_data.loc[train_index], vrai_data.loc[test_index]

        X_train = full_pipeline.fit_transform(data_train) # Attention ici on utilise la mÃ©thode fit_transforme
        y_train = np.array(data_train["grav_class"])

        X_test = full_pipeline.transform(data_test)  # et ici on utilise la mÃ©thode transforme
        y_test = np.array(data_test["grav_class"])

        model = SGDClassifier(max_iter=1000, loss='hinge')
        model.fit(X_train, y_train)
        ac_score.append(accuracy_score(y_test, model.predict(X_test)))

        #print(accuracy_score(y_test, model.predict(X_test)))

    print("****************")
    print("PrÃ©cision moyenne : ", np.mean(ac_score))
    return np.mean(ac_score)
```
