# 1.Système de recommandation simpliste
#### Recommandation en fonction de la note moyenne : Doit-on donné la même importance à deux films ayant la même note mais avec un nombre de votant considérablement différent ?
#### On va plutôt chercher à normaliser la note moyenne par rapport au nombre de votant, c'est à dire donner plus d'importance au note ayant le plus de votant.
#### Formule de la note pondérée :
<img src="img\norm_imdb.png" width="600">

- r la note moyenne
- v le nombre de vote
- m le nombre minimum de vote pour apparaitre dans la liste
- c la valeur moyenne du nombre de vote

## On calcule le nombre de votes recueillis par le film du 80ème centile

#### Quel est la note de film qui me permet de séparer les 20% des films les mieux noté des 80% le moins bien noté ?

```python
m = movies_metadata['vote_count'].quantile(0.80)
movies_metadata['vote_count'].quantile(0.80)
```

#### On ne concerve que les films de plus de 45 min et de moins de 300 minutes

```python
q_film = movies_metadata[(movies_metadata['runtime'] >= 45) & (movies_metadata['runtime'] <= 300)]
```

#### On ne concerve que le films qui on recueillis plus de m votes (on veut que le film soit populaire pour être recommandé)

```python
q_film = movies_metadata[(movies_metadata['vote_count'] >= m)]
```

```python
q_film.shape
```

#### On calcule le score c

```python
c = movies_metadata['vote_average'].mean()
c
```

#### On va enfin pouvoir calculer la note pondéré

```python
def note_ponderee(x, m=m, c=c):
    v = x['vote_count']
    r = x['vote_average']
    return ((v/(v+m)*r)+(m/m+v)*c)
```

```python
q_film['score'] = q_film.apply(note_ponderee, axis=1)
```

```python
q_film = q_film.sort_values('score', ascending=False)
les30film_a_recomander = q_film[['title', 'vote_count', 'vote_average', 'score', 'runtime']].head(30)
```

#### Recommandation basé sur les connaissances

```python
def recom_conn(gen_movies, percentile=0.80):
    genre = input("Entrez votre genre de film préféré : ")
    low_time = int(input("Saisissez la durée de film minimal : "))
    high_time = int(input("Saisissez la durée de film maximal : "))
    low_year = int(input("Saisissez l'année de film minimal : "))
    high_year = int(input("Saisissez l'année de film maximal : "))
    
    movies_reco = gen_movies.copy()
    
    movies_reco = movies_reco[(movies_reco['genre']==genre) &
                              (movies_reco['runtime']>=low_time) &
                              (movies_reco['runtime']<=high_time) &
                              (movies_reco['year']>=low_year) &
                              (movies_reco['year']<=high_year)
                             ]
    
    c = movies_reco['vote_average'].mean()
    m = movies_reco['vote_count'].quantile(percentile)
    
    q_film = movies_reco[(movies_reco['vote_count'] >= m)]
    
    q_film['score'] = q_film.apply(lambda x: (x['vote_count']/(x['vote_count']+m)*x['vote_average'])
                                   +(m/m+x['vote_count'])*c
                                   , axis=1)
    
    q_film = q_film.sort_values('score', ascending=False)
    q_film = q_film[['title', 'vote_count', 'vote_average', 'score', 'runtime']].head(30)
    
    return q_film
```

```python
recom_conn(gen_movies)
```

#### On va vectoriser et appliquer la transformation TF-IDF sur la colonne 'overview'

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

#On ramplace les NaN par une chaine vide
movies['overview'] = movies['overview'].fillna('')
# Vectorisation et trnasformation tf-idf des commentaires
tfidf_matrix = tfidf.fit_transform(movies['overview'])
tfidf_matrix.shape
```

#### Utilisation de la formule de similitude cosinus pour déterminer la similtude des films sur la base de leurs descriptions
<img src="img\cosine_sim.png" width="600">
Retourne un score compris entre -1 et 1. Plus le Score est élevé et plus les document sont similaire.

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Je construit une cartographie inverse des index et des titres des films, supprime les doublons
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()
```

```python
indices
```

#### Je crée ma fonction qui prend en entré le titre d'un film et retourne les recommandation

```python
#exemple avec le film 'Jumanji' ayant l'indice 1
sim_scores = list(enumerate(cosin_sim[1]))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
# le film qui lui ressemble le plus d'unpoint de vue description est celui avec l'indice 21633.
sim_scores
```

```python
# Je met tout ça en fonction pour que tout cela soit plus propre
def recom_con(title, cosine_sim=cosine_sim, movies=movies, indices=indices):
    # Je récupère l'indice du film
    idx = indices[title]
    # Je récupère tout les score de similitude du film concerné avec tout les films 
    sim_scores = list(enumerate(cosin_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Je récupère l'indice des 10 films les plus proches
    sim_scores = sim_scores[1:11]
    movies_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movies_indices]
```
